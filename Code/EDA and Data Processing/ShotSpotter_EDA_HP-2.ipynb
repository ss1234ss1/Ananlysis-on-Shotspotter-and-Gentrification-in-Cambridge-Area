```python
def common_columns(*dfs):
    # Start with the columns of the first DataFrame
    common_cols = set(dfs[0].columns)

    # Find the intersection of columns for all DataFrames provided
    for df in dfs[1:]:
        common_cols = common_cols.intersection(set(df.columns))

    return list(common_cols)

# Call the function with three DataFrames
common_cols = common_columns(df_excel2, df_excel3, df_excel4)
print("Common columns across all three datasets:", common_cols)
```

```python
len(common_cols)
len(df_excel2.columns)
len(df_excel3.columns)
len(df_excel4.columns)
```

```python
columns_df1 = set(df_excel2.columns)
columns_df2 = set(df_excel3.columns)
columns_df3 = set(df_excel4.columns)




common_columns = columns_df1 & columns_df2 & columns_df3
all_columns = columns_df1 | columns_df2 | columns_df3

non_common_columns = all_columns - common_columns

print("Columns that are not common across all files:", non_common_columns)
```

```python
# Work on only Cambridge city
df_csv_2=df_csv[df_csv["owner_city"]=="Cambridge"]
```

```python
# Display the first few rows of relevant columns to confirm their contents
print(df_csv_2[['site_addr', 'addr_str', 'addr_num']].head())
```

```python
# Check if 'total_value' represents the housing price you want to analyze
# Group by the street name and calculate the average housing price
avg_price_by_street = df_csv_2.groupby('addr_str')['total_value'].mean().sort_values(ascending=False)

# Display the results
print(avg_price_by_street)
```

```python
avg_price_by_street = avg_price_by_street[avg_price_by_street > 0]
top_10_streets = avg_price_by_street.head(10)
print(top_10_streets)
```

```python
import numpy as np

log_transformed_prices = np.log1p(avg_price_by_street)
log_transformed_prices.plot(kind='hist', bins=30, figsize=(10, 6))
plt.title('Log-Transformed Distribution of Average Property Values by Street')
plt.xlabel('Log of Average Property Value')
plt.ylabel('Frequency')
plt.show()
```

```python
print(avg_price_by_street.describe())
```

```python
import matplotlib.pyplot as plt

top_10_streets.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Streets by Average Property Value')
plt.xlabel('Street')
plt.ylabel('Average Property Value')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by zip code and calculate the average assessed value
avg_assessed_value_by_zip = df_csv_2.groupby('owner_zip')['total_value'].mean()

# Display the results
print(avg_assessed_value_by_zip)

# You can also plot this data if you'd like to visualize it
# For example, using a bar plot:
avg_assessed_value_by_zip.plot(kind='bar', figsize=(12, 6))
plt.title('Average Assessed Value by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Average Assessed Value')
plt.xticks(rotation=45)
plt.show()
```
